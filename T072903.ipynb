```python
import random
import matplotlib.pyplot as plt

days = list(range(1,32))


#celsius = []
#for _ in range(31):
#    celsius.append(round(random.uniform(25,45),ndigits=2))

#list comprehension

#celsius = [round(random.uniform(25,45),ndigits=2) for x in range(31) if x >15]
celsius = [round(random.uniform(25,45),ndigits=2) for x in range(31)]
figure = plt.figure(figsize=(8,4))
axes = figure.add_subplot(1,1,1)
axes.plot(days,celsius,'r--.')
axes.grid(axis='y')
axes.set_xticks(days)
plt.show()
```
